```python
df_suk = df_immigration.loc[['Spain', 'United Kingdom'], range(2009, 2014)]

df_suk
```

```python
df_suk = df_suk.transpose()

df_suk.plot(kind = 'line')

plt.title ('Immigration Numbers to Spain and U.K.')
plt.ylabel ('Number of Immigrants')
plt.xlabel ('Years')

plt.show()
```

```python
# ndf_suk is the new data frame for spain and the uk
# here we are standardising the values of the new data frame against the maxiumum immigration number

ndf_suk = df_suk
ndf_suk[['Spain']] = df_suk[['Spain']] / df_suk[['Spain']].max()
ndf_suk[['United Kingdom']] = df_suk[['United Kingdom']] / df_suk[['United Kingdom']].max()

ndf_suk
```

```python
ndf_suk.plot(figsize = (12, 8))

plt.title ('Immigration Ratios to Spain and U.K.')
plt.ylabel ('Ratio of Immigrants (Yearly Immigration / Maximum Yearly Imigration)')
plt.xlabel ('Years')

plt.text(2012.08, 0.83, 'Lowest Point')

plt.show()
```

## Area Plots

```python
df_immigration.head()
```

```python
df_immigration['Total'] = df_immigration.sum(axis = 1)
df_immigration.head()
```

```python
df_immigration.sort_values(['Total'], ascending = False, axis = 0, inplace = True)

df_top3 = df_immigration.head(3)
df_top3.head()
```

```python
df_top3 = df_top3[range(2009, 2014)].transpose()

df_top3.head()
```

```python
df_top3.plot(kind = 'area', stacked = False, figsize = (15, 10))

plt.title('Top 3 Immigration Trends')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
# alpha is the transparency value and it is 0.5 by default

df_top3.plot(kind = 'area', alpha = 0.1, stacked = False, figsize = (15, 10))

plt.title('Top 3 Immigration Trends')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
df_top3.plot(kind = 'area', alpha = 0.1, figsize = (15, 10))

plt.title('Top 3 Immigration Trends')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
df_germany.plot(kind = 'bar', figsize = (12, 10), rot = 45)

plt.xlabel('Year')
plt.ylabel('Immigration Numbers')
plt.title('Immigration to Germany 2009 - 2018')

plt.annotate('jump',
            xy = (5.9, 1.45*10**6),
            xytext = (4.9, 0.8*10**6), 
            xycoords = 'data',
            arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3', color = 'lightblue', lw = 2))

plt.show()
```

```python
df_germany.plot(kind = 'bar', figsize = (12, 10), rot = 45)

plt.xlabel('Year')
plt.ylabel('Immigration Numbers')
plt.title('Immigration to Germany 2009 - 2018')

plt.annotate('jump',
            xy = (5.9, 1.45*10**6),
            xytext = (4.9, 0.8*10**6), 
            xycoords = 'data',
            arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3', color = 'lightblue', lw = 2))

plt.annotate('fall',
            xy = (7, 0.9*10**6),
            xytext = (5.9, 1.48*10**6), 
            xycoords = 'data',
            arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3', color = 'lightgreen', lw = 2))

#We are annotating the arrows with some text
plt.annotate('Immigration Jump 2014 - 2015',
             xy = (5, 0.9*10**6),
             rotation = 75.7,
             va = 'bottom', 
             ha = 'left')

plt.annotate('Immigration Fall 2015 - 2016',
             xy = (6.2, 1.45*10**6),
             rotation = -70.5,
             va = 'top', 
             ha = 'left')

plt.show()
```
